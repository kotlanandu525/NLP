# **Stemming**

stemming reduces words to their root form by removing prefixes and suffixes


likes--> like


liked--> like


likely --> like

```python
from nltk.stem import PorterStemmer
#porterStemer is a basic so drawbacks exists
porter_stemmer=PorterStemmer()
words=['likes','liked','running','happily']
print("before stemming")
print(words)
stemmed_words=[porter_stemmer.stem(word) for word in words]
print(stemmed_words)
```

```python
#Snoewball stemmer is an enhanced version of the porter stemmer supports multple languages
from nltk.stem import SnowballStemmer
snowball_stemmer=SnowballStemmer(language='english')
wors_to_stem=['running','jumped','happily','programming']
print("before stemming")
print(wors_to_stem)
stemmed_words=[snowball_stemmer.stem(word) for word in wors_to_stem]
print(stemmed_words)
```

```python
#landcaster stemmer is known for being more agggresive and faster than other stemmers

from nltk.stem import LancasterStemmer
lancaster_stemmer=LancasterStemmer()
words_to_stem=['running','jumped','happily','programming']
print("before stemming")
print(words_to_stem)
stemmed_words=[lancaster_stemmer.stem(word) for word in words_to_stem]
print(stemmed_words)
```

## ## **Lemmatization**

Lemmatization produces meaningful root words.

**Example**

better → good

running → run

Lemmatization is an important text pre-processing technique in Natural Language Processing (NLP) that reduces words to their base form known as a "lemma." For example, the lemma of "running" is "run" and "better" becomes "good."

Unlike stemming which simply removes prefixes or suffixes, it considers the word's meaning and part of speech (POS) and ensures that the base form is a valid word. This makes lemmatization more accurate as it avoids generating non-dictionary words.

**Used for:**

Improves accuracy

Reduced Data Redundancy

Better NLP Model Performance





```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize('running',pos='v'))
print(lemmatizer.lemmatize('better',pos='a'))
words=['running','jumped','happily','programming']
print("before lemmatization")
print(words)
lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in words]
print(lemmatized_words)
```

## **Bag of Words (BoW)**

BoW represents text as word frequency vectors.

In Natural Language Processing (NLP) text data needs to be converted into numbers so that machine learning algorithms can understand it. One common method to do this is Bag of Words (BoW) model. It turns text like sentence, paragraph or document into a collection of words and counts how often each word appears but ignoring the order of the words. It does not consider the order of the words or their grammar but focuses on counting how often each word appears in the text.

This makes it useful for tasks like text classification, sentiment analysis and clustering.


```python
import re
dataset=[]

text={
    "i LoVe coding!",
    ' i love python@#',
    'python coding is    fun'
}

for sentence in text:
  sentence=sentence.lower() #lowercase
  sentence=re.sub(r'[^\w\s]','',sentence) #remove punctuation
  dataset.append(sentence)

print(dataset)
```

```python
#Counting word frequencies

from collections import defaultdict
word_freq=defaultdict(int)
for sentence in dataset:
  words=sentence.split()
  for word in words:
    word_freq[word]+=1
print(word_freq)
```

```python
#Select most frequent words

import heapq
freq_words=heapq.nlargest(3,word_freq,key=word_freq.get)
print(freq_words)
#
```

# **step4:Build bag of Words Matrix**
Now we create a matrix:

Rows=sentences

Columns=frequent words

value=1(present) or 0

```python
x=[]
for sentence in dataset:
  vector=[]
  words=sentence.split()
  for word in freq_words:
    if word in words:
      vector.append(1)
    else:
      vector.append(0)
  x.append(vector)
print(x)
```


# **N-grams**

N-grams capture word sequences.

N-gram is a technique that splits text into sequences of N consecutive words to capture context and word order for machine learning models.

N-gram = group of N consecutive words

1 word → Unigram

2 words → Bigram

3 words → Trigram

#### **✅ Simple Example**

Sentence:

#### "I love coding"
####🔹 Unigrams (N = 1)
["I", "love", "coding"]
####🔹 Bigrams (N = 2)
["I love", "love coding"]
####🔹 Trigrams (N = 3)
["I love coding"]


```python
#step1 preprocessing the text

import re
text={
    'i love coding',
    'i love python ',
    'python coding is    fun'
}

dataset=[]

for sentence in text:
  sentence=sentence.lower()
  sentence=re.sub(r'[^\w\s]','',sentence)
  dataset.append(sentence)

print(dataset)
```

```python
#step 2 crete N-grams

def generate_ngrams(sentence,n):
  words=sentence.split()
  ngrams=[]
  for i in range(len(words)-n+1):
    ngrams.append("".join(words[i:i+n]))
  return ngrams
```

```python
#example bigram(n=2)
for sentence in dataset:
  bigrams=generate_ngrams(sentence,2)
  print(bigrams)
print("????")
#example Trigrams(N=3)
for sentence in dataset:
  trigrams=generate_ngrams(sentence,3)
  print(trigrams)
#
```

```python
#step3:Count N-gram frequencies
from collections import defaultdict
ngram_freq=defaultdict(int)
for sentence in dataset:
  unigrams=generate_ngrams(sentence,2)
  for gram in unigrams:
    ngram_freq[gram]+=1
print(ngram_freq)
#
```

```python
#Build matrix
x=[]
for sentence in dataset:
  vector=[]
  sentence_ngrams=generate_ngrams(sentence,2)
  for gram in freq_words:
    if gram in sentence_ngrams:
      vector.append(1)
    else:
      vector.append(0)
  x.append(vector)
print(x)
```
